# Step 1b
Define utility functions
and clear cache

```python
# standard_deviations = 0 #@param {type:"slider", min:0, max:3, step:0.1}
from torchvision import transforms
import requests
import gc

images_per_chunk = 40 #@param {type:"integer"}

def get_image(url_or_path):
    if 'https://' in url_or_path: # UNUSED
        r = requests.get(url_or_path, stream=True)
        img = Image.open(r.raw)
    else:
        img = Image.open(url_or_path)
    img = img.convert("RGB") # png alpha channels screw it up
    return img

@functools.lru_cache()
def to_tensor(filepath):
    '''Return the device-specific transformed tensor for the filepath or url'''
    img = get_image(filepath)
    img = transforms.functional.to_tensor(img)
    img = transforms.functional.resize(img, (224,224))
    img = img.unsqueeze(0)
    return img.to(device)

def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

def imagePathsInDir(directory, recursive=False):
    '''Return a list of paths to images in the given directory'''
    image_exts = ['jpg', 'png', 'jfif']
    def iterator():
        for ext in image_exts:
            if recursive:
                yield from glob.glob(os.path.join(directory, '**', f'*.{ext}'), recursive=True)
            else:
                yield from glob.glob(os.path.join(directory, f'*.{ext}'), recursive=False)
    return list(iterator())

def rsample(list_, max_):
    return random.sample(list_, min(max_, len(list_)))


def relpath(p, parent):
    return p[len(parent):].replace('\\', '/')

def getTensorScores(file_paths):
    acc = []
  
    # Construct smart iterator
    iterator = [file_paths]
    if len(file_paths) > images_per_chunk:
        iterator = tqdm.tqdm([*chunks(file_paths, images_per_chunk)], unit=f'{images_per_chunk}-file chunk')
    
    for sublist in iterator:
        if not sublist:
            # Sometimes sublist is empty? Don't process in that case
            continue
        
        gc.collect()
        torch.cuda.empty_cache()
    
        tensors = torch.cat([*map(to_tensor, sublist)], 0)
        with torch.no_grad():
            next_ = clip_model.encode_image(tensors)
            acc.append(next_)
  
    if acc == []:
        return None
    return torch.cat(acc)

print(f"Done. {images_per_chunk=}")
```

```python
model_basedir = '/content/models' #@param {type:"string"}

max_sample_images = images_per_chunk * 4 #@param {type:"integer"}

model_whitelist = ['*']
model_blacklist = ['unsorted']
model_force_recompute = []

model_dirs_recursive = True #@param {type:"boolean"}

quality_dirs = sum([glob.glob(f'{model_basedir}/{m}/') for m in model_whitelist], [])

dont_reprocess = True #@param {type:"boolean"}
#@markdown Skip existing models, so you can add a model directory without recomputing everything. Set to false to do a full recompute.

# Build image encoding models for each "model" folder
pbar = tqdm.tqdm(quality_dirs, unit="model dir")
try:
    for quality in pbar:
        qname = os.path.basename(os.path.dirname(quality))
        pbar.set_description(qname)

        if qname.startswith('maybe_') or qname.lower() in model_blacklist:
            print(f"Skipping {qname}")
            continue

        if qname not in quality_dict or (not dont_reprocess) or qname in model_force_recompute:
            # image_features ~= clip_model.encode_image
            paths = imagePathsInDir(quality, recursive=model_dirs_recursive)
            image_paths = rsample(paths, max_sample_images)
            try:
                image_features = getTensorScores(image_paths)
                if image_features is not None:
                    image_features /= image_features.norm(dim=-1, keepdim=True)
                    quality_dict[qname] = image_features
            except (FileNotFoundError):
                import traceback
                traceback.print_exc()
                continue
except KeyboardInterrupt:
    pass

print()
pprint({k: hash(v) for k, v in quality_dict.items()})
```

# Step 3: Rate each image

```python
# Clear quality results

quality_results = []
```

```python
topk_points = 12 #@param {type:"integer"}
#@markdown Base each image's confidence on its distance between the closest 
#@markdown `topk_points` vectors.

max_images = 1000 #@param {type:"integer"}
#@markdown Process first `max_images` image files in image_dir

use_spherical_distance_algo = False #@param {type:"boolean"}
recursive_image_search = True #@param {type:"boolean"}
#@markdown Match files in subdirectories?

image_dir = '/content/images/' #@param {type:"string"}
# image_dir = os.path.join(model_basedir, 'unsorted')
# image_dir = os.path.join(model_basedir)

from torch.nn import functional as F
def spherical_distance(x, y):
    x = F.normalize(x, dim=-1)
    y = F.normalize(y, dim=-1)
    l = (x - y).norm(dim=-1).div(2).arcsin().pow(2).mul(2)
    return l 

skip_processed = True
# If you add models in step 2, leaving this on may produce unwanted results!

if skip_processed:
    processed_paths = {r['path'] for r in quality_results}

image_paths = imagePathsInDir(image_dir, recursive=recursive_image_search)
for image_path in tqdm.tqdm(image_paths[:max_images], unit="image"):
    if skip_processed and image_path in processed_paths:
        continue
        
    result = {"path": image_path}

    # Build our "imageset" for comparison (just one image) and get features
    image_features = getTensorScores([image_path])

    # Compute similarity against each quality
    for qname, qfeatures in quality_dict.items():
        # scores is an array of distances between vectors
        if use_spherical_distance_algo:
            # need to reverse-sort these values
            scores = spherical_distance(image_features, qfeatures)
            result[qname] = sys.maxsize - torch.topk(scores, min(len(scores), topk_points)).values.mean()
        else:
            scores = torch.nn.CosineSimilarity()(image_features, qfeatures)
            result[qname] = torch.topk(scores, min(len(scores), topk_points)).values.mean()

    quality_results.append(result)

print("Done.")
```

# Step 4: Group images

```python
low_score_threshhold = 0.79 #@param {type:"slider", min:0, max:1, step:0.01}
#@markdown If not at least this confident (%), assign the special "LOW" category.

# No special box to clear this variable because this calculation is very cheap

buckets = {
    k: []
    for k in quality_dict.keys()
}
buckets['LOW'] = []

for result in quality_results:
  # filter out path, just tensors
  scores = {k: v for k, v in result.items() if not isinstance(v, str)}

  # sort by tensor, then grab largest
  (best_q, b_score) = sorted(scores.items(), key=lambda s: s[1])[-1]

  # If the score is this low, maybe it doesn't fit any of them?  
  if b_score < low_score_threshhold:
    best_q = 'LOW'

  # if not relpath(result['path'], image_dir).startswith(best_q):
  result['bucket'] = best_q
  buckets[best_q].append(result)

print({bk: len(bv) for bk, bv in buckets.items()})
time.sleep(1)
```
